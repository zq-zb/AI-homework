```python
import os
import chardet

# 你的文本文件所在目录（如 project5/data）
text_dir = "project5/data/"

# 遍历所有 .txt 文件
for filename in os.listdir(text_dir):
    if filename.endswith(".txt"):
        file_path = os.path.join(text_dir, filename)
        with open(file_path, "rb") as f:
            raw_data = f.read()
            result = chardet.detect(raw_data)
        # 只输出编码不是GBK的文件（重点关注这些文件）
        if result["encoding"] != "ascii" and result["confidence"] > 0.8:
            print(f"文件：{filename} → 编码：{result['encoding']}（置信度：{result['confidence']:.4f}）")
```
